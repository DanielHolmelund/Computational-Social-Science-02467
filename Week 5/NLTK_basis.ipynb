```python
from nltk.corpus import brown
cfd = nltk.ConditionalFreqDist(
                (genre, word)
                for genre in brown.categories()
                for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

> * 15.Write a program to find all words that occur at least three times in the Brown Corpus.

> * 16.Write a program to generate a table of lexical diversity scores (i.e. token/type ratios), as we saw in 1.1. Include the full set of Brown Corpus genres (nltk.corpus.brown.categories()). Which genre has the lowest diversity (greatest number of tokens per type)? Is this what you would have expected?

```python
diversity_dict = {}
for category in brown.categories():
    tokens = len(brown.words(categories = category))
    types = len(set(brown.words(categories = category)))
    score = format(types / tokens, ".3f")
    diversity_dict[category] = score

print(diversity_dict)
```

The lowest category of diveristy is "learned", which is a bit surprising, since you would hope that in a learning enviroment the students are exposed to a wide variey of subject and language. This low level of lexical diversity might be attributed to a lot of standard formulations and structures that both students and teachers apply. 

> * 17.Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
from nltk.corpus import stopwords
def top_50_freq_words(text):
    fdist = FreqDist(word.lower() for word in text if word.lower() not in stopwords.words("english") and word.isalpha())
    return fdist.most_common(50)
top_50_freq_words(text1)
```

> * 18.Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
def top_50_freq_bigrams(text):
    bigrams = list(nltk.bigrams(text)) #list to unpack the iterator object
    stopwords_ = stopwords.words("english")
    fdist = FreqDist(bigram for bigram in bigrams if bigram[0] not in stopwords_ and
                    bigram[1] not in stopwords_ and bigram[0].isalpha() and bigram[1].isalpha())
    return fdist.most_common(50)
top_50_freq_bigrams(text1)
```
