## Explain in your own words what aspect of language lexical diversity describes. 

Lexical diversity describes how rich the vocabulary of the given book is because it calculates the fraction of unique words to the total number of words. However, one drawback of the lexical diversity measure is that it does not take into account that larger text has a much greater number of tokens but the available words within a language stays more or less constant, hence, it can be misleading to compare texts with different lengths.

◑ Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

emma = nltk.corpus.gutenberg.words('austen-emma.txt')

def n_most_common(text, n):
    fdist = nltk.FreqDist(text)
    most_frequent = []
    for word in fdist:
        if word not in stop_words and word.isalnum():
                most_frequent.append(word)
    return most_frequent[:n]

print(n_most_common(emma, 50))
```

◑ Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
def remove_stop(text):
    filtered_text = []
    for word in fdist:
        if word not in stop_words and word.isalnum():
            filtered_text.append(word)
    return filtered_text

def n_most_freq_bigrams(text, n):
    text_filtered = remove_stop(text)
    temp = nltk.bigrams(text_filtered)
    fdist = nltk.FreqDist(temp)
    return fdist.most_common(n)

print(n_most_freq_bigrams(emma, 50))
```
